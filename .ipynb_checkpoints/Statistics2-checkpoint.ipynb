# Deal with missing data

```python
dates = pd.date_range('2005-12-31', '2013-12-07')
symbols = ['SPY', 'FAKE2']
df = m.get_data(symbols, dates)
m.plot_data(df)
```

```python
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)
m.plot_data(df, title="Stock Prices filled")
```

```python
def fill_missing_values(df_data):
    df_data.fillna(method='ffill', inplace=True)
    df_data.fillna(method='bfill', inplace=True)
    return df_data
```

```python
symbols = ['JAVA', 'FAKE1', 'FAKE2']
dates = pd.date_range('2005-12-31', '2014-12-07')
df = m.get_data(symbols, dates)
df = fill_missing_values(df)
m.plot_data(df)
```

# Histograms of daily_returns

```python
dates = pd.date_range('2009-01-01', '2012-12-31')
symbols = ['SPY']
df = m.get_data(symbols, dates)
m.plot_data(df)
daily_returns = s.daily_returns(df)
m.plot_data(daily_returns, title='DailyReturns')
```

```python
daily_returns.hist(bins=20)
mean = daily_returns['SPY'].mean()
print('Mean:', mean)
std = daily_returns['SPY'].std()
print('Std:', std)
kurt = daily_returns['SPY'].kurt()
print('Kurtosis:', kurt, 'fat_tails' if kurt > 0 else 'skinny_tails')
plt.axvline(mean, color='w', linestyle='dashed', linewidth=2)
plt.axvline(mean-std, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean+std, color='r', linestyle='dashed', linewidth=2)
```

# Scatter plots

```python
dates = pd.date_range('2009-01-01', '2012-12-31')
symbols = ['SPY', 'XOM', 'GLD']
df = m.get_data(symbols, dates)
m.plot_data(df)
#Compute daily_rets
daily_returns = s.daily_returns(df)
```

```python
#Scatterplot SPY vs XOM
daily_returns.plot(kind='scatter', x='SPY', y='XOM', edgecolors='b')
#Linear Regression
beta_XOM, alpha_XOM = np.polyfit(daily_returns['SPY'], daily_returns['XOM'], 1)
y = beta_XOM*daily_returns['SPY'] + alpha_XOM
plt.plot(daily_returns['SPY'], y, '-', color='r')
```

```python
#Scatterplot SPY vs GLD
daily_returns.plot(kind='scatter', x='SPY', y='GLD', edgecolors='b')
#Linear Regression
beta_GLD, alpha_GLD = np.polyfit(daily_returns['SPY'], daily_returns['GLD'], 1)
y = beta_GLD*daily_returns['SPY'] + alpha_GLD
plt.plot(daily_returns['SPY'], y, '-', color='r')
```

# Portfolio Statistics

```python
date_range = pd.date_range('2014-04-01', '2014-11-01')
symbols = ['AAPL', 'IBM']
df = m.get_data(symbols, date_range)
```

```python
def print_statistics(df):
    cum_ret = s.cumulative_return(df)
    print("Cumulative Return")
    print(cum_ret.head(n=3))
    print("Daily Return")
    daily_rets = s.daily_returns(df)
    m.plot_data(daily_rets)
    print(daily_rets.head(n=3))
    print("Average Daily Return:")
    print(s.average_daily_return(df))
    print("Risk:")
    print(s.risk(df))
    print("Sharpe_Ration:") 
    print(s.sharpe_ratio_dayily(daily_rets, 0))
print_statistics(df)
```

```python
#Dollars
start_val = 100000
dates = pd.date_range('2017-01-02', '2017-06-20')
print(dates.size)
symbols = ['AAPL', 'XOM', 'GLD']
df = m.get_data(symbols, dates)
symbols = ['AAPL', 'XOM', 'GLD']
allocs = [0.4, 0.4, 0.2]

#1
prices = df.ix[:, symbols]
#2
normed = m.normalize_data(prices)
print(normed.head())
#3
alloced = normed * allocs
print(alloced.head())
#4
pos_val = alloced * start_val
print(pos_val.head())
#5
portfolio_val = pos_val.sum(axis=1)
print("Portfolio")
print(portfolio_val.head())

m.plot_data(portfolio_val, title="Portfolio")
```

```python
cum_ret = s.cumulative_return(portfolio_val)
print("Cumulative Return")
print(cum_ret.head(n=3))
print("Daily Return")
daily_rets = s.daily_returns_simple(portfolio_val)
m.plot_data(daily_rets, title="Daily_returns")
print(daily_rets.head(n=3))
print("Average Daily Return:")
print(daily_rets.mean())
print("Risk:")
print(daily_rets.std())
print("Sharpe_Ration:") 
print(s.sharpe_ratio_dayily(daily_rets, 0))
```
