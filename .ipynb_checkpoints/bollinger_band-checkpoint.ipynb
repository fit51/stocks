```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import requests
import json
import scipy
import seaborn as sns
import datetime
import util.methods as m
import util.statistics as s
import util.portfolio as p
import scipy.optimize as spo
```

```python
url = "https://api.tinkoff.ru/trading/stocks/list?start=0&end=2000&sortType=ByEarnings&orderType=Desc&country=Foreign"
list_resp = json.loads(requests.get(url).text)
```

```python
list_tickers = [x['symbol']['ticker'] for x in list_resp['payload']['values']]
list_names = [x['symbol']['showName'] for x in list_resp['payload']['values']]
```

```python
tickers_data = []
for ticker in list_tickers:
    history = "https://api.tinkoff.ru/trading/symbols/historical_data?ticker=" + ticker + \
    "&from=2016-06-21T21%3A00%3A00.000Z&to=2017-12-22T17%3A56%3A50.523Z&resolution=D"
    data = json.loads(requests.get(history).text)
    tickers_data.append((ticker, data))
```

```python
window = 50
bollinger_window = 15

bollinger_candidates = []
for i in range(len(tickers_data)):
    cur_ticker = tickers_data[i]
    ticker = cur_ticker[0]
    values = np.array([x[1] for x in cur_ticker[1]['payload']['data']])
    values = values / values[0]
    avg = np.zeros(values.shape)
    avg[window - 1:]= np.convolve(values, np.ones((window,)) / window, mode='valid')
    avg[:window]= avg[window - 1]
    std = np.std(values)
    low_band = avg - std * 2
    hi_band = avg + std * 2
    show_name = list_names[i]
    down_crossed_exists = np.sum(values[-bollinger_window:] - low_band[-bollinger_window:] < 0) > 0
    up_crossed_exists = values[-1] - low_band[-1] > 0
    bollinger_mark = np.argmax(values[-bollinger_window:] - low_band[-bollinger_window:] < 0) 
    if down_crossed_exists and up_crossed_exists:
        bollinger_candidates.append((ticker, values, avg, low_band, hi_band, bollinger_window - bollinger_mark, show_name))
```

```python
plt.figure(figsize=(10, 5 * len(bollinger_candidates)))
for i, cand in enumerate(bollinger_candidates):    
    sub = plt.subplot(len(bollinger_candidates), 1, i + 1)
    ticker, values, avg, low_band, hi_band, b_mark, show_name = cand
    y_mark = list(range(len(values) - b_mark, len(values)))
    ax = sub.axes
    ax.set_title("%s. Ticker:(%s)" % (show_name, ticker), fontsize=20)
    sub.plot(avg, 'r')
    sub.plot(values, 'g')
    sub.plot(low_band, 'b')
    sub.plot(hi_band, 'b')
    sub.plot(y_mark, values[-b_mark:],'black', linewidth=3.0)
plt.show()
```

```python
def time(s):
    return datetime.datetime.fromtimestamp(int(s)).strftime('%Y-%m-%d')

candidates = [b[0] for b in bollinger_candidates]
print(candidates)

full_data = {}
stop_date = 2555360320.0 #2050 year
start_date = 0.0
for d in tickers_data:
    for c in candidates:
        if c == d[0]:
            print(c, time(d[1]['payload']['data'][0][0]), time(d[1]['payload']['data'][-1][0]))
            if start_date < d[1]['payload']['data'][0][0]: start_date = d[1]['payload']['data'][0][0]
            if stop_date > d[1]['payload']['data'][-1][0]: stop_date = d[1]['payload']['data'][-1][0]
            full_data[c] = [(time(dd[0]), dd[1]) for dd in d[1]['payload']['data']]
print(len(full_data))
print(time(start_date), time(stop_date))
```

```python
def sharpe_ratio_allocs(allocs, normed, rfr=0.0, sf=252.0):
    alloced = normed * allocs
    port_val = alloced.sum(axis=1)
    daily_rets = s.daily_returns_simple(port_val)[1:]
    std_daily_ret = daily_rets.std()
    daily_rfr = (1.0 + rfr)**(1/sf) - 1
    sharpe_ratio = sf ** (1/2) * ((daily_rets - daily_rfr).mean() / std_daily_ret)
    return sharpe_ratio

def sharpe_ratio_error(allocs, normed, rfr=0, sf=252):
    return sharpe_ratio_allocs(allocs, normed, rfr, sf) # mulyiply -1 if sharpe_ratio is positive

labels = ['Date', 'Adj Close']
dates = pd.date_range(time(start_date), time(stop_date))
df = pd.DataFrame(index=dates)
for c in full_data:
    df_temp = pd.DataFrame.from_records(data=full_data[c], columns=labels, index="Date")
    df_temp=df_temp.rename(columns={'Adj Close': c})
    df=df.join(df_temp)
    
n = len(full_data)
allocs_equal = [1/n] * n
prices = df.dropna()
normed = m.normalize_data(prices)
print("Initial Sharpe_ration={}".format(sharpe_ratio_allocs(allocs_equal, normed)))

allocs_bounds = [(0, 1)] *n
constraints = ({ 'type': 'eq', 'fun': lambda allocs: 1.0 - np.sum(allocs) })
min_result = spo.minimize(sharpe_ratio_error, allocs_equal, args=(normed,), method='SLSQP', \
                          bounds=allocs_bounds, \
                          constraints=constraints, options={'disp': True})
allocs = min_result.x
print("Optimal Allocs = {}".format(allocs))
print("Func value = {}".format(sharpe_ratio_allocs(allocs, normed)))
```
